# Introduction to Functions

A function is some expression where each input (x) has exactly 1output (y)

A vertical line test can be used to determine if an expression is a function or not.

```python
import matplotlib.pyplot as plt
import numpy as np
import math

def plot_graph(graph_type, equation, range_):
    x = np.linspace(*range_, num=1000)
    y = None
    if graph_type == 'cubic':
        y = x**3
    elif graph_type == 'cube root':
        y = x**(1/3)
    elif graph_type == 'parabola':
        y = x**2
    elif graph_type == 'custom':
        y = eval(equation)
    else:
        raise ValueError('Invalid graph type')

    plt.plot(x, y)
    plt.show()
```

```python
plot_graph('parabola', None, (-10, 10))
```

```python
plot_graph('cube root',None, (0, 10))
```

```python
plot_graph('custom',"1/x", (1, 10))
```

The above are all functions.

### Example

The below expression is a circle

$$
x^2 + y^2 = 25
$$

The equation above represents the equation of a circle in two-dimensional space. In general, the equation $x^2 + y^2 = r^2$ represents a circle with a center at the origin (0, 0) and a radius r. The value of $r^2$ determines the size of the circle, so in this case, the circle has a radius of 5 units.

The equation $x^2 + y^2 = 25$ states that the distance of every point on the circle from the origin (0, 0) is equal to 5 units. This means that the points on the circle are equidistant from the origin, which is why it forms a circular shape. The graph of this equation is a circle centered at the origin with a radius of 5 units.

However this is not a function.

Simplifying for y gives us two ± answers for a given x value. 
$$
y = ±√(25 - x^2)
$$

However, we can break this down into two separate functions. $f(x)$ and $g(x)$. Where,
$$
f(x) = √(25 - x^2)
$$
$$
g(x) = -√(25 - x^2)
$$

$f(x)$ and $g(x)$ are essentially the top and bottom halves of a circle. 

# Piecewise Functions

The function changes depending on the input value. It is defined in different ways over different intervals

A simple example is the absolute value function. 

$$
f(x) = |x|
$$
$$
f(-5) = |-5| = 5
$$

The domain of the function is defined as
$$
f(x) = 
  {x   if  x ≥ 0
  {-x  if  x < 0
$$

You can graph any piecewise function by graphing each piece individually. 

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    #the piecewise function f(x) is defined using np.piecewise with two intervals: x < 0 and x >= 0. 
    #The functions to apply for each interval are lambda x: -x and lambda x: x, respectively
    return np.piecewise(x, [x < 0, x >= 0], [lambda x: -x, lambda x: x])

x = np.linspace(-10, 10, 1000)
y = f(x)

plt.plot(x[x >= 0], y[x >= 0], color='green', label='x >= 0')
plt.plot(x[x < 0], y[x < 0], color='red', label='x < 0')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of f(x) = |x| in piecewise form')
plt.legend()
plt.grid(True)
plt.show()
```

>In the code used to graph f(x) = |x| in piecewise form (above), lambda functions are used as arguments to the np.piecewise function. The np.piecewise function takes a series of intervals and the corresponding functions to apply to each interval. The lambda functions are used to define the corresponding functions to apply to each interval.
For example, lambda x: -x is a lambda function that takes x as an argument and returns the value -x. This lambda function is used to define the corresponding function to apply to the interval x < 0, so that f(x) = -x for x < 0. Similarly, lambda x: x is a lambda function that takes x as an argument and returns the value x. This lambda function is used to define the corresponding function to apply to the interval x >= 0, so that f(x) = x for x >= 0.
So in this case, the lambda functions are used to define the piecewise function f(x) = |x| in a compact and easily understandable way, without having to define separate functions for each interval.

### Example

$$
g(x) = 
  {0   if  x ≤ 1
  {√(1-x^2) if -1<x<1
  {x  if  x ≥ 1
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_piecewise_function(x_min, x_max, num_points, color1, color2, color3):
    x = np.linspace(x_min, x_max, num_points)
    f = np.piecewise(x, [x <= -1, (x > -1) & (x < 1), x >= 1], [0, lambda x: np.sqrt(1 - x**2), lambda x: x])

    plt.plot(x[x <= -1], f[x <= -1], color=color1, label='x <= 1')
    plt.plot(x[(x > -1) & (x < 1)], f[(x > -1) & (x < 1)], color=color2, label='-1 < x <= 1')
    plt.plot(x[x >= 1], f[x >= 1], color=color3, label='x >= 1')

    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Piecewise Function g(x)')
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
plot_piecewise_function(-2, 2, 1000, 'red', 'green', 'blue')
```

# Domain & Range

Domain: All the values you can input into a function that gives a valid real number output. 

Range: All the output values. 

## Domain Restrictions

- The square root function has a restriction on its domain, as it is undefined for negative values:
$$
√x, x ≥ 0
$$

- The logarithm function has a restriction on its domain, as it is undefined for values less than or equal to 0:
$$
log(x), x > 0
$$

- The inverse tangent function has a restriction on its domain, as it has multiple outputs for a given input and the range of the function is limited to (-π/2, π/2):
$$
tan^-1(x), x ∈ (-∞,∞)
$$

- The reciprocal function has a restriction on its domain, as it is undefined for x = 0:
$$
1/x, x ≠ 0
$$

## Natural Domain

All values that work in the formula. 
It asks the question, for a given formula, is there anything that you cannot plug into that?
The natural domain of a function is usually expressed as a set of numbers, or as a range of values. The exact range of values depends on the function and the mathematical constraints imposed on its inputs. For example, the natural domain of the square root function is the set of all non-negative real numbers.

For example in the function $f(x) = x^3$, the natural domain is:

D: ALL REAL NUMBERS

or expressed another way, $f: x ∈ R$

where $x ∈ R$ means "x belongs to the set of all real numbers (R)". This notation is commonly used in statistics and mathematics to denote the domain of a function, and it is read as "x is an element of the set of all real numbers".

### Example 2

$$
h(x) = tan(x)
$$

Put another way, $tan(x)$ is equal to, 
$$
tan(x) = \frac{sin(x)}{cos(x)}
$$

In this case, the $cos(x)$ is in the denominator, therefore $cos(x)\neq0$.The $cos(x)$ function spits out zero with $\frac{\Pi}{2}$, $\frac{3\Pi}{2}$...etc.

Therefore the domain, $x\neq \frac{\Pi}{2}$, $\frac{3\Pi}{2}$...


```python
import numpy as np
import matplotlib.pyplot as plt

# Define x values. This will create a plot of the cosine function over the range from -π to π.
x = np.linspace(-4*np.pi, 4*np.pi, 200)

# Compute y values for cosine function
y = np.cos(x)

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the cosine function
ax.plot(x, y)

# Add a title and labels for x and y axes
ax.set_title("Cosine Function")
ax.set_xlabel("x")
ax.set_ylabel("y")

# Add a grid
ax.grid(True)

# Find the x values where y is closest to 0. The np.isclose() function is used to find the x values where y is closest to 0 (within a tolerance of 0.01)
zeros = x[np.isclose(y, 0, atol=0.01)]

# Add labels for the points where y=0
for zero in zeros:
    ax.text(zero, 0, f"{zero:.2f}", ha="center", va="center")

# Display the plot
plt.show()
```

### Example 3

 $$
 f(x) = \sqrt{x^2-5x+6}
 $$
 


Remember the radicant $\geq0$. 

But first, factor the radicant. 

To factor the expression $x^2 - 5x + 6$, you can follow these steps:

1. Write down the expression in the form $ax^2 + bx + c$, where $a$, $b$, and $c$ are coefficients. In this case, $a = 1$, $b = -5$, and $c = 6$.

2. Find two numbers that multiply to $a*c$ (in this case, $1*6 = 6$) and add up to $b$ (in this case, $-5$). The two numbers are $-2$ and $-3$, since $-2 * -3 = 6$ and $-2 + (-3) = -5$.

3. Rewrite the middle term -5x as the sum of the two numbers found in step 2: $x^2 - 2x - 3x + 6$.

4. Group the first two terms and the last two terms together and factor out the greatest common factor from each group: $x(x - 2) - 3(x - 2)$.

5. Notice that $(x - 2)$ is a common factor of both terms, so factor it out: $(x - 2)(x - 3)$.

6. Verify the factored expression by multiplying the factors to see if you get the original expression: $(x - 2)(x - 3) = x^2 - 5x + 6$.

Therefore, the factored form of $x^2 - 5x + 6$ is $(x - 2)(x - 3)$.

$$
(x-3)(x-2)\geq0
$$

The interval between 2 and 3 is NOT in the domain. Inputs between these two numbers will give a negative in the radicant. 
The domain of this function is the set of all real numbers x such that x^2 - 5x + 6 >= 0.

Domain: $x ∈ (-∞, 2] ∪ [3, ∞)$

This notation indicates that the domain includes all real numbers less than or equal to 2 and greater than or equal to 3, but excludes the interval between 2 and 3. This is because the expression under the square root is negative when x is between 2 and 3, which is not allowed in the domain.

```python
def ex3_f(x):
    return math.sqrt(x**2 - 5*x + 6)
```

### Example 4

 $$
 f(x) = \frac{x^2-4}{x-2}
 $$

Remember the denominator $x-2\neq0$
therefore,
$x\neq0$

Can we cancel out the domain problem by simplifying the expression? No, domain problems cannot be cancelled out by simplification. The original domain remain set for the expression. However, for the purposes of graphing, if the domain of an expression cancels out during simiplification, then there is a hole in the graph (removable discontiunity), else you get an asymptote.  

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function
def f(x):
    return (x**2 - 4)/(x - 2)

# Define the x values for the plot
x = np.linspace(-10, 10, 1000)

# Plot the function
plt.plot(x, f(x))

# Add a vertical line for the removable discontinuity at x = 2
plt.axvline(x=2, color='red', linestyle='--')

# Set the x and y limits for the plot
plt.xlim(-10, 10)
plt.ylim(-10, 10)

# Add labels for the x and y axes
plt.xlabel('x')
plt.ylabel('y')

# Add a title for the plot
plt.title('Graph of f(x) = (x^2 - 4)/(x - 2)')

# Display the plot
plt.show()
```
